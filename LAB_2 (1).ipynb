```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Display Image
def display_img(img):
  img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  plt.imshow(img_RGB);
```

## Write code to load an image, display it on screen, and write it to a file in a different format.

```python
im = cv2.imread('flower.jpg', cv2.IMREAD_COLOR)
print(im.shape)
plt.imshow(im);
```

```python
im2 = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
print(im2.shape)
plt.imshow(im2);
```

```python
cv2.imwrite("myImg.jpg", im2)
```

## Access the 100  100-pixel value from the image's top-left corner and change its color to green.

```python
cropped_image = im[0:100, 0:100]
cropped_image2 = cv2.cvtColor(cropped_image,cv2.COLOR_BGR2RGB)
print(im2.shape)
plt.imshow(cropped_image2);
```

## [3] write code for drawing lines, and rectangles on a canvas, as shown in Fig. 1.

```python
line_image = cv2.line(im, (0,0), (4288,2848), (0, 255, 0), 9)
display_img(line_image)
```

```python
rect_image = cv2.rectangle(im, (1000,500), (3500,2500), (0, 0, 255), 9,)
display_img(rect_image)
```

## [4] Write code for drawing concentric circles with a minimum and maximum radius of 0 and 150cm, respectively, which increment by 25 at each step.

```python
for i in range(0, 1400, 25):
  cir_image = cv2.circle(im, (2000,1500), i, (255, 0, 0), 9)
display_img(cir_image)
```

## [5] draw 25 random circles, with a radius value in the range [5, 200) with different shapes and colors in a single canvas.

```python
import random

# Generate 25 random tuples
random_tuples = []
for _ in range(25):
  x = random.randint(3000, 3900)
  y = random.randint(300, 2400)
  random_tuples.append((x, y))

for x, y in random_tuples:
    radius = random.randint(5, 200)
    cir_rand_image = cv2.circle(im, (x, y), radius, (255,255,255), 8)

display_img(cir_rand_image)
```
